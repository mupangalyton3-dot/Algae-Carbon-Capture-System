Capture ≈ (CO₂ × Light × Time) / Scaling factor

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Create dataset
np.random.seed(42)

co2 = np.random.uniform(100, 1000, 100)       # CO2 concentration
light = np.random.uniform(50, 300, 100)       # Light intensity
time = np.random.uniform(1, 10, 100)          # Time

# Simulated CO2 capture formula
capture = (co2 * light * time) / 100000 + np.random.normal(0, 2, 100)

# Create DataFrame
data = pd.DataFrame({
    "CO2": co2,
    "Light": light,
    "Time": time,
    "Capture": capture
})

print(data.head())

plt.scatter(data["CO2"], data["Capture"])
plt.xlabel("CO2 Concentration")
plt.ylabel("CO2 Captured")
plt.title("CO2 vs Capture (Algae System)")
plt.show()

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

# Features and target
X = data[["CO2", "Light", "Time"]]
y = data["Capture"]

# Split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Model
model = LinearRegression()
model.fit(X_train, y_train)

# Predictions
predictions = model.predict(X_test)

print("Model Trained Successfully")

from sklearn.metrics import r2_score

score = r2_score(y_test, predictions)
print("Model Accuracy (R²):", score)
```

CO₂ capture increases with:
>
Higher CO₂ concentration
>
Increased light intensity
>
Longer exposure time
